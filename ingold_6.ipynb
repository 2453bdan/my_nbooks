```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("avocado.csv")
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
filtered_df = df.query("year == 2015 and type == 'conventional' and region == 'Albany'")[["Date", "Total Bags", "Small Bags"]]
filtered_df.head()
```

```python
melted_df = pd.melt(df, id_vars="Date", value_vars=["Total Bags", "Small Bags"], var_name="Bags", value_name="Count")
display(melted_df)
```

```python
sns.lineplot(data=melted_df, x="Date", y="Count", hue="Bags")
```

```python
df["Sales Volume"] = pd.qcut(df["Total Volume"], 4, labels=["Poor", "Modest", "Good", "Excellent"])
df.head()
```

```python
sns.countplot(data=df, x="year", hue="Sales Volume")
```

```python
melted_bags = pd.melt(df, id_vars=["year"], value_vars=["Small Bags", "Large Bags", "XLarge Bags"], var_name="Bags", value_name="Count")
display(melted_bags)
```

```python
sns.relplot(data=melted_bags, kind="line", x="year", y="Count", hue="Bags")
```
